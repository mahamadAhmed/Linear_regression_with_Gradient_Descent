```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(200)
delta = np.random.uniform(-50,30, size=(200,))

y = .4 * x +3 + delta

plt.scatter(x, y)
plt.show()
```

```python
train_size = int(len(x) * 0.8)
train_x, train_y = x[:train_size], y[:train_size]

test_x, test_y = x[train_size:], y[train_size:]
```

```python
import random
w = 0
b = 0
learning_rate = 0.0001
epochs = 1000
```

```python
#mean squared error (MSE) loss function
def mse_loss(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)
```

```python

# Define gradient descent function
def gradient_descent(x, y_true, w, b, learning_rate, epochs):
    # lists to store loss and parameters
    losses = []
    weights = []
    biases = []

    for i in range(epochs):
        # Calculate y_pred and loss
        y_pred = w * x + b
        loss = mse_loss(y_true, y_pred)
        losses.append(loss)
        
        # Calculate gradients
        dw = np.mean((y_pred - y_true) * x)
        db = np.mean(y_pred - y_true)

        # Update weights and biases
        w -= learning_rate * dw
        b -= learning_rate * db
        weights.append(w)
        biases.append(b)

    return losses, weights, biases

losses, weights, biases = gradient_descent(train_x, train_y, w, b, learning_rate, epochs)

```

```python
print("Final weights: ", weights[-1])
print("Final biases: ", biases[-1])

# predictions on test set
test_pred = weights[-1] * test_x + biases[-1]


# Calculate MSE on test set
accuracy = mse_loss(test_y, test_pred)
print("accuracy:", accuracy)
```
